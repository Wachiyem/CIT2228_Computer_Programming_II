Example 2: Computing factorial of n.
1. Start with result= 1.
2. Multiply result by each number from 1 to n.
3. Output result

```python
n = 5
result = 1
for i in range(1, n + 1):
    result *= i
print(f"The factorial of {n} is: {result}")
```

## Python Code: Average Height
1. Input the heights
2. average = sum of heights divide total number

```python
ariel = 148
gitau = 175
mercy = 163

sum_height = ariel + gitau + mercy

average = sum_height/ 3

print(f"Average height of {ariel}, {gitau}, and {mercy} is: ", average)
```

```python
import statistics

heights = [148, 175, 163]

average_height = statistics.mean(heights)

print(average_height)
```

## Sequential flow example: Ohm's Law

$V = IR$

$V$ = voltage,
$I$ = current,
$R$ = resistance

```python
I = 10 # current
R = 200 # resistance
V = I * R # Voltage

print("Voltage = ", V,'ohms')
```

## Selection Example(`if-else`)

```python
body_temp = float(input('Enter body temp in celcius: '))

if body_temp > 37:
  print('fever detected')
elif body_temp < 0:
  print('frozen body')
else:
  print('normal body temp')
```
